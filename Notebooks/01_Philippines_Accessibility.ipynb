# Case Study 1: Philippines Non-Profit Expansion


## Objective
Identify areas in the Philippines where a global non-profit could potentially expand its reach using Kiva loan locations.


## Import Required Libraries

Import GeoPandas and supporting libraries for geospatial analysis and visualization.

```python
import geopandas as gpd
```

```python
DATA_PATH = r"C:\Data Analytics\GeoSpatialAnalysis"
```

```python
loans_filepath = DATA_PATH + r"\Data\kiva_loans\kiva_loans\kiva_loans.shp"
```

## Load Kiva Loan Data

Load the geospatial dataset containing Kiva loan locations around the world.

```python
world_loans=gpd.read_file(loans_filepath)
```

```python
world_loans.head()
```

```python
import geopandas as gpd

url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"

world = gpd.read_file(url)

world.head()
world.shape
```

## Visualize Global Loan Distribution

Plot Kiva loan locations on a world map to understand their overall geographic distribution.

```python
ax=world.plot(figsize=(18,18),color='whitesmoke',edgecolor='lightgrey',linestyle=':')
world_loans.plot(markersize=2,ax=ax)
```

## Focus on the Philippines

Filter the dataset to include only loan records from the Philippines.

```python
PHL_loans=world_loans.loc[world_loans.country=='Philippines']
```

```python
PHL_loans.shape
```

## Load Philippines Boundary

Import the administrative boundary of the Philippines using a KML file.

```python
import fiona
fiona.supported_drivers["KML"] = "rw"

PHL = gpd.read_file(
    r"C:\Data Analytics\GeoSpatialAnalysis\Data\Philippines_AL258.kml",
    driver="KML"
)
```

```python
PHL.shape
```

## Identify Potential Expansion Areas

Visualize loan locations within the Philippines to identify regions that may benefit from expanded non-profit outreach.

```python
import matplotlib.pyplot as plt
ax=PHL.plot(figsize=(13,13),color='whitesmoke',edgecolor='lightgrey')
PHL_loans.plot(markersize=3,ax=ax)

plt.savefig(
    r"C:\Data Analytics\GeoSpatialAnalysis\images\philippines_expansion_map.png",
    dpi=300,
    bbox_inches="tight"
)
```

## Conclusion

Based on the spatial distribution of Kiva loan locations, several regions in the Philippines appear underserved. Mindoro, a relatively large island in the central Philippines, stands out due to the absence of loans in the current dataset. This suggests that Mindoro could be a promising candidate for further investigation and potential expansion of Field Partner operations.
